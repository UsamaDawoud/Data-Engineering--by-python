```python
# some countries is very small in areas , so  will filled with zero all it's NULL  if the 'Agriculture' is null 
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Region has zero Null , so it can be usefull to fill the missing data depending on the 'mean' of each Region 
```

```python
# for climate 
df['Climate']= df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

```python
# for Literacy (%) 
df['Literacy (%)']= df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

```python
#for Net migration 
df['Net migration']= df['Net migration'].fillna(df.groupby('Region')['Net migration'].transform('mean'))
```

```python
#for Birthrate
df['Birthrate']= df['Birthrate'].fillna(df.groupby('Region')['Birthrate'].transform('mean'))
```

```python
# for Net migration
df['Net migration']= df['Net migration'].fillna(df.groupby('Region')['Net migration'].transform('mean'))
```

```python
#For Infant mortality (per 1000 births)
df['Infant mortality (per 1000 births)']= df['Infant mortality (per 1000 births)'].fillna(df.groupby('Region')['Infant mortality (per 1000 births)'].transform('mean'))
```

```python
df.isnull().sum()
```

```python
# Can do the same for other but in some countries it different from thier nieghbors so to drop it will be right chose 
```

```python
df = df.dropna()
```

# univariate

```python
from sklearn.impute import  SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan , strategy='mean')

# Missing_values : int, float, str, np.nan or None, default=np.nan , The placeholder for the missing values. All occurrences of
#   missing_values` will be imputed. For pandas' dataframes with nullable integer dtypes with missing values, 'missing_values'
#   should be set to `np.nan`, since `pd.NA` will be converted to `np.nan`. 
#
# strategy : (mean, median or most frequent) 
#
# (e.g. impute.SimpleImputer)
# univariate, which imputes values in the i-th feature dimension using only non-missing values in that feature dimension 
```

```python
df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']] = imp.fit_transform(
    df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']])
                                                                                                     
```

```python
df.info()
```

```python
df.head(25)
```

```python
# for the strings columns it is better to deal with minualy or simply delet it
# but as feature available it possible to deal with autmatically
imp1_2 = SimpleImputer(missing_values=np.nan , strategy='most_frequent')

df[['species','island','sex']] = imp1_2.fit_transform( df[['species','island','sex']])
   
```

# multivariate imputation algorithms

```python
# multivariate imputation algorithms use the entire set of available feature dimensions to estimate the missing values
#  (e.g. impute.IterativeImputer).
```

```python
df2 = pd.read_csv('mod_penguins_size.csv')
```

```python
df2.head(25)
```

```python
df2.info()
```

```python
# Note This estimator is still experimental for now: the predictions and the API might change without any deprecation cycle. 
# To use it,there is a  need to explicitly import enable_iterative_imputer:
#
# A more sophisticated approach is to use the IterativeImputer class, which models each feature with missing values as a
# function of other features, and uses that estimate for imputation.
#
# initial_strategystr, default=mean --  {mean, median, most_frequent, or constant}.
```

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_iterative_imputer  # noqa
# now you can import normally from sklearn.impute
from sklearn.impute import IterativeImputer
```

```python
imp2 = IterativeImputer()
```

```python
df2[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']] = imp2.fit_transform(
    df2[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']])
```

```python
df2.head(25)
```

```python
imp2_2 = IterativeImputer(missing_values=np.nan , initial_strategy='most_frequent')

df2[['species','island','sex']] = imp2_2.fit_transform(df2[['species','island','sex']])
```

# KNNImputer

```python
# The KNNImputer class provides imputation for filling in missing values using the k-Nearest Neighbors approach. 
```

```python
from sklearn.impute import KNNImputer
knn_imp = KNNImputer(n_neighbors=2, weights= "uniform")
# - 'uniform' : uniform weights. All points in each neighborhood are  weighted equally.
#     
#    - 'distance' : weight points by the inverse of their distance.
#      in this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.
#      
#    - callable : a user-defined function which accepts an array of distances, and returns an array of the same 
#      shape containing the weights.
      
```

```python
df3 = pd.read_csv('mod_penguins_size.csv')
```
